## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)    # use plt.show()  for non jupyter IDE
plt.show()
```

```python
plt.plot(x,y,'r--')   # for dotted red color graph
```

```python
plt.plot(x,y)
plt.title('Demo title')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.subplot(1,2,1)     # to create subplot in plot with argument ( no. of rows, no. of columns, subplot number)
plt.plot(x,y, 'r')
plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

### Object Oriented method

```python
fig = plt.figure()    # creates figure object ( as blank figure )
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])    # craeted an axes of given size
axes.plot(x,y)  # plot the x and y data in the axes
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title(' demo title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])   # left , bottom, width , height
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])   # creates subplot (20%, 50%, 40%,30%) of original
#axes2 = fig.add_axes([0.4, 0.1, 0.5,0.5])   # to move the right bottom
axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_xlabel('x-axis')
axes1.set_ylabel('y-axis')
axes1.set_title('big graph title')

axes2.set_xlabel('x-axis')
axes2.set_ylabel('y-axis')
axes2.set_title('small graph title')
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols =2)  # creates nxm subplots

#axes is an array of objects of axis

for currentAxis in axes:
    currentAxis.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols =2)
axes[0].plot(x,y)
axes[1].plot(x,y)   # same as above

plt.tight_layout()   # to avoid any overlapping in plot
```

### Figure size and DPI

```python
fig = plt.figure(figsize=(3,4), dpi = 100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows =1, ncols =2, figsize = (3,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# save a figure 
fig.savefig('savedFig.png', dpi = 200)
```

```python
fig, axes = plt.subplots()

axes.plot(x,y, label = 'squared')
axes.plot(y,x, label = 'square root')
axes.legend(loc = 0)  # loc = 0 is for best location
#axes.legend(loc = 10)  # loc = 0 is for center location
#axes.legend(loc = (0.1,0.1))  # loc = (0.1,0.1) is for 10% from left 10% from bottom
plt.tight_layout()
```

### Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF8C00', lw = 5 ,alpha = 0.5, ls ='--')   # HEX color, lw is for linewidth, alpha is for opacity, ls for line style
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF8C00', marker ='o',markersize = 16,markerfacecolor = 'yellow', markeredgewidth =5, markeredgecolor ='green')

# markers to mark the values of dataset
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF8C00')

ax.set_xlim([-10,10])   # set limits of x axis
ax.set_ylim([-10,30])   # set limits of y axis
```
