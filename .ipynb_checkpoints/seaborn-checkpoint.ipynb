# Seaborn

```python
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')    # preinstalled dataset
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])  #  gives distribution graph
```

```python
sns.distplot(tips['total_bill'], kde=False)   # removes curve and plot is just like a histogram
```

```python
sns.distplot(tips['total_bill'],kde= False, bins = 30)   # will giev more bars
```

```python
sns.jointplot(x ='total_bill', y='tip' , data = tips)   # give joint figure
```

```python
sns.jointplot(x ='total_bill', y='tip' , data = tips, kind = 'hex')   # creates hex colored density figure
```

```python
sns.jointplot(x ='total_bill', y='tip' , data = tips, kind = 'reg')  # creates a reegression line
```

```python
sns.jointplot(x ='total_bill', y='tip' , data = tips, kind = 'kde')
```

```python
sns.pairplot(tips)    # created pairwise scatter plot of each adat and histogram for itself
```

```python
sns.pairplot(tips, hue = 'sex')   # to use categorical value as well
```

```python
sns.pairplot(tips, hue='sex', palette ='autumn')    # to set a color pallete
```

```python
sns.rugplot(tips['total_bill'])    # kinds a density of x values   used by histogram
```

```python
sns.kdeplot(tips['total_bill'])   # to get only kernel density estimation curve
```

## Categorical data

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data =tips)
```

```python
sns.barplot(x='sex', y='total_bill', data =tips, estimator =np.std)   # to estimate plot in the basis of standard deviation
```

```python
sns.countplot(x = 'sex', data = tips)   # counts number of males and females
```

```python
sns.boxplot(x='day', y='total_bill', data = tips)
```

```python
sns.boxplot(x='day', y='total_bill', data = tips, hue = 'smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data = tips)
```

```python
sns.violinplot(x='day', y='total_bill', data = tips, hue ='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data = tips, hue ='sex', split = True)
```

```python
sns.stripplot(x='day', y ='total_bill', data = tips)
```

```python
sns.stripplot(x='day', y ='total_bill', data = tips, jitter = True)   # for understanding better clarity of density
```

```python
sns.stripplot(x='day', y ='total_bill', data = tips, hue = 'sex')
```

```python
sns.stripplot(x='day', y ='total_bill', data = tips, hue = 'sex', dodge = True)   # dodge means split
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)    # kind a comination of voilin and strip plot

# not recommended for large dataset
```

```python
# if you are using swarm plot use voilin plot stacked above

sns.violinplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.catplot(x = 'day', y ='total_bill',data = tips, kind='bar')
```

```python
sns.catplot(x = 'day', y ='total_bill',data = tips, kind='violin')
```

## Matrix Plot

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
print(tc)
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot = True)
```

```python
sns.heatmap(tc, annot = True, cmap='coolwarm')
```

```python
flights.head()
```

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values='passengers')
```

```python
fp.head()
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap = 'coolwarm')
```

```python
sns.heatmap(fp, cmap = 'coolwarm', linecolor = 'white', linewidth =1)
```

```python
sns.heatmap(fp, cmap = 'coolwarm', linecolor = 'black', linewidth =3)
```

```python
sns.clustermap(fp)    # clusters similar type of features from data ( more related are grouped to each other)
```

```python
sns.clustermap(fp, cmap='coolwarm')
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale = 1)   # normalize the data
```
